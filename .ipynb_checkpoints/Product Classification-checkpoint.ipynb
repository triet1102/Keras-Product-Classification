```python
from keras import models
from keras import layers
from keras.utils import to_categorical, plot_model
import os
import shutil
import cv2
from keras.models import Sequential, Model
from keras import Input
import pydot
import keras as K
import keras
from keras.utils.vis_utils import model_to_dot
keras.utils.vis_utils.pydot = pydot
import matplotlib.pyplot as plt
```

```python
base_dir_path = '/home/triet/Desktop/test_qopius/Implementation of Classification Model/data2'
```

```python
products_dir = sorted([o for o in os.listdir(base_dir_path)])
```

```python
for file in os.listdir(os.path.join(base_dir_path,products_dir[0])):
    im = cv2.imread(os.path.join(base_dir_path,products_dir[0], file))
    print(im.shape)
```

```python
data_split_folder = ['train', 'validation', 'test']
for f in data_split_folder:
    if os.path.exists(os.path.join(base_dir_path, f)):
        shutil.rmtree(os.path.join(base_dir_path, f))
    os.mkdir(os.path.join(base_dir_path, f))
```

```python
data_split_folder_product = ['p1', 'p2', 'p3', 'p4', 'p5', 'p6']
for f1 in data_split_folder:
    for f2 in data_split_folder_product:
        os.mkdir(os.path.join(base_dir_path, f1, f2))
```

```python
for f1 in data_split_folder:
    for f3,f4 in (tuple(zip(products_dir, data_split_folder_product))):
        if f1 == 'train':
            idx1, idx2 = 0, 50
        elif f1 == 'validation':
            idx1, idx2 = 50, 60
        else:
            idx1, idx2 = 60, 90
        for file in os.listdir(os.path.join(base_dir_path, f3))[idx1:idx2]:
            src = os.path.join(base_dir_path, f3, file)
            dst = os.path.join(base_dir_path, f1, f4, file)
            shutil.copyfile(src, dst)
```

```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True)
```

```python
test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
train_generator = train_datagen.flow_from_directory(
    os.path.join(base_dir_path,data_split_folder[0]),
    target_size=(150, 150),
    batch_size=20,
    class_mode='categorical')
```

```python
train_generator = train_datagen.flow_from_directory(
    os.path.join(base_dir_path,data_split_folder[0]),
    batch_size=20,
    class_mode='categorical')
```

```python
validation_generator = test_datagen.flow_from_directory(
    os.path.join(base_dir_path,data_split_folder[1]),
    batch_size=5,
    class_mode='categorical')
```

```python
test_generator = test_datagen.flow_from_directory(
    os.path.join(base_dir_path,data_split_folder[2]),
    batch_size=5,
    class_mode='categorical')
```

```python
def block1(inputs):
    l = layers.BatchNormalization()(inputs)  
    l = layers.Activation('relu')(l)
    l = layers.Conv2D(32, 3, padding='same')(l)  
    l = layers.BatchNormalization()(l)
    l = layers.Activation('relu')(l)
    l = layers.Conv2D(32, 3, padding='same')(l)
    x = layers.add([l,inputs])
    return x
```

```python
def block2(inputs):
    l = layers.BatchNormalization()(inputs)
    l = layers.Activation('relu')(l)
    l = layers.Conv2D(32, 3, padding='same')(l)
    l = layers.BatchNormalization()(l)
    l = layers.Activation('relu')(l)
    l = layers.Conv2D(32, 3, padding='same')(l)
    r = layers.Conv2D(32, 3, padding='same')(inputs)
    r = layers.BatchNormalization()(r)
    x = layers.add([l,r])
    return x
```

```python
input_tensor = Input(shape=(None,None,3))
x = layers.Conv2D(32, 3)(input_tensor)
x = layers.BatchNormalization()(x)
x = layers.Activation('relu')(x)
x = layers.MaxPooling2D()(x)
l = layers.Conv2D(32, 3)(x)
l = layers.BatchNormalization()(l)
l = layers.Activation('relu')(l)
l = layers.Conv2D(32, 3)(l)
r = layers.Conv2D(32, 5)(x)
r = layers.BatchNormalization()(r)
x = layers.add([l,r])
#x = block1(x)
#x = block2(x)
#x = block1(x)
#x = block2(x)
#x = block1(x)
#x = block2(x)
#x = block1(x)
x = layers.BatchNormalization()(x)
x = layers.Activation('relu')(x)
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dense(256)(x)
output_tensor = layers.Dense(6, activation='softmax')(x)
```

```python
model = Model(input_tensor, output_tensor)
```

```python
model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['acc'])
```

```python
model.summary()
```

```python
history = model.fit_generator(
    train_generator,
    steps_per_epoch=15,
    epochs=20,
    validation_data=validation_generator,
    validation_steps=12)
```

```python
test_loss, test_acc = model.evaluate_generator(test_generator, steps=36)
```

```python
print(test_loss, test_acc)
```

```python
history.history.keys()
```

```python
val_loss = history.history['val_loss']
val_acc = history.history['val_acc']

train_loss = history.history['loss']
train_acc = history.history['acc']

epochs = range(len(val_loss))
```

```python
plt.plot(epochs, train_acc, 'bo', label='training accuracy')
plt.plot(epochs, val_acc, 'b', label='validation accuracy')
plt.legend()
plt.title('training and validation accuracy')

plt.figure()

plt.plot(epochs, train_loss, 'bo', label='training loss')
plt.plot(epochs, val_loss, 'b', label='validation loss')
plt.legend()
plt.title('training and validation loss')

plt.show()
```
