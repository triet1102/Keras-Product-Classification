```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True)
```

```python
test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(150, 150),
    batch_size=20,
    class_mode='categorical')
```

```python
validation_generator = test_datagen.flow_from_directory(
    validation_dir,
    target_size=(150, 150),
    batch_size=5,
    class_mode='categorical')
```

```python
test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(150, 150),
    batch_size=5,
    class_mode='categorical')
```

```python
#
#
#
#
```

```python
def block1(inputs):
    l = layers.BatchNormalization()(inputs)  
    l = layers.Activation('relu')(l)
    l = layers.Conv2D(32, 3, padding='same')(l)  
    l = layers.BatchNormalization()(l)
    l = layers.Activation('relu')(l)
    l = layers.Conv2D(32, 3, padding='same')(l)
    x = layers.add([l,inputs])
    return x
```

```python
def block2(inputs):
    l = layers.BatchNormalization()(inputs)
    l = layers.Activation('relu')(l)
    l = layers.Conv2D(32, 3, padding='same')(l)
    l = layers.BatchNormalization()(l)
    l = layers.Activation('relu')(l)
    l = layers.Conv2D(32, 3, padding='same')(l)
    r = layers.Conv2D(32, 3, padding='same')(inputs)
    r = layers.BatchNormalization()(r)
    x = layers.add([l,r])
    return x
```

```python
input_tensor = Input(shape=(150,150,3))
x = layers.Conv2D(32, 3)(input_tensor)
x = layers.BatchNormalization()(x)
x = layers.Activation('relu')(x)
x = layers.MaxPooling2D()(x)
l = layers.Conv2D(32, 3)(x)
l = layers.BatchNormalization()(l)
l = layers.Activation('relu')(l)
l = layers.Conv2D(32, 3)(l)
r = layers.Conv2D(32, 5)(x)
r = layers.BatchNormalization()(r)
x = layers.add([l,r])
x = block1(x)
x = block2(x)
#x = block1(x)
#x = block2(x)
#x = block1(x)
#x = block2(x)
#x = block1(x)
x = layers.BatchNormalization()(x)
x = layers.Activation('relu')(x)
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dense(256)(x)
output_tensor = layers.Dense(6, activation='softmax')(x)
```

```python
model = Model(input_tensor, output_tensor)
```

```python
model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['acc'])
```

```python
model.summary()
```

```python
history = model.fit_generator(
    train_generator,
    steps_per_epoch=20,
    epochs=10,
    validation_data=validation_generator,
    validation_steps=12)
```

```python
test_loss, test_acc = model.evaluate_generator(test_generator, steps=36)
```

```python
print(test_loss, test_acc)
```
